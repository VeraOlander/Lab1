```python
import matplotlib.pyplot as plt
```

```python
# full code for one sequence to create a separate dictionary and bar chart for each single sequence
# take a DNA sequence which is initially a string, convert it into lower string, than to a list and sort the list
# count the letters in each element of the list and convert to a dictionary
# create a bar chart for each sequence/dictionary


def dict_for_each_sequence(N):
    sequence=items_edited[N-1]
    sequence_lower=sequence.lower()
    list_sequence=[]
    for i in sequence_lower:
        list_sequence.append(i)
    list_sequence.sort()
    result = {}
    for letter in list_sequence:
        if letter not in result:
            result[letter] = 0
        result[letter] += 1
    print(f"DNA sequence {N}: {result}")
    DNA_letters=result.keys()
    Frequency=result.values()
    plt.bar(DNA_letters,Frequency)
    plt.title(f"DNA sequence {N}")
    plt.xlabel("DNA letters")
    plt.ylabel("Frequency")
    plt.show() # iteration did not work w/o this line
    
    
    return result
```

```python
# iterate over all sequences from the file

number_of_sequences=len(items_edited)

for i in range(1,number_of_sequences+1):
    dict_for_each_sequence(i)
```
