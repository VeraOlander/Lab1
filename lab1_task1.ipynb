read the text file dna_raw.txt


```python
with open("dna_raw.txt", "r") as f:
    dna_raw = f.read()
    
print(dna_raw)
```

```python
# create a list of items from the text
items = dna_raw.split(">")

print(items)
```

```python
# remove the first element that doesn't contain any sequence data
items.remove('')
print(items)
```

```python
# create a dictionary of sequences from the list of items
dict_of_seq={}
for item in items:
    key, value = item.strip().split("\n")
    dict_of_seq[key.strip()] = value.strip()

print(dict_of_seq)
```

```python
# create a list of sequences which are the dictionary's values
list_of_values=list(dict_of_seq.values())
print(list_of_values)
```

```python
# full code for one sequence to create a separate dictionary and bar chart for each single sequence
# take a DNA sequence which is initially a string, convert it into lower string, than to a list and sort the list
# count the letters in each element of the list and convert to a dictionary
# create a bar chart for each sequence/dictionary


def dict_for_each_sequence(N):
    sequence=list_of_values[N-1]
    sequence_lower=sequence.lower()
    list_sequence=[]
    for i in sequence_lower:
        list_sequence.append(i)
    list_sequence.sort()
    result = {}
    for letter in list_sequence:
        if letter not in result:
            result[letter] = 0
        result[letter] += 1
    print(f"DNA sequence {N}: {result}")
    DNA_letters=result.keys()
    Frequency=result.values()
    plt.bar(DNA_letters,Frequency)
    plt.title(f"DNA sequence {N}")
    plt.xlabel("DNA letters")
    plt.ylabel("Frequency")
    plt.show() # iteration did not work w/o this line
    
    
    return result
   

   
dict_for_each_sequence(1) # just to to test that it works for a single sequence
```
